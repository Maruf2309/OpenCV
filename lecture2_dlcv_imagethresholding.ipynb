### What is Image Thresholding ?

Image Thresholding is a technique in image processing that convert grayscale into binary image. This is done by selecting a threshold value (or mutliple threshold values) and classifying the pixel in the image based intensity relative to this threshold

**Applicaion of Image Thresholding in Computer Vision**
1. Object Detection
2. Image Segmentation
3. Medical Imaging
4. Face Detection
5. Motion Detection
6. Quality Control

```python
# Loading Image
image_path = '/content/orange_1.jpg'
img = cv.imread(image_path, cv.IMREAD_GRAYSCALE)

# cv.threshold(gray_img, threshold_val, max_val, threshold_type)
# ret,thresh1 = cv.threshold(img, 190,255, cv.THRESH_BINARY)

# reset
ret, thresh1 = cv.threshold(img, 127,255, cv.THRESH_BINARY)
ret, thresh2 = cv.threshold(img, 127,255, cv.THRESH_BINARY_INV)
ret, thresh3 = cv.threshold(img, 127,255, cv.THRESH_TRUNC)        # >threshold = threshold; <= unchanged
ret, thresh4 = cv.threshold(img, 127,255, cv.THRESH_TOZERO)       # >threshold unchanged; <= 0(black);
ret, thresh5 = cv.threshold(img, 127,255, cv.THRESH_TOZERO_INV)   # >threshold = 0(black); <= unchanged

titles = ['Original','Binary','BINARY_INV','TRUNC', 'TOZERO','TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]

for i in range(6):
  plt.subplot(2, 3, i+1), plt.imshow(images[i], 'gray', vmin=0, vmax=255)
  plt.title(titles[i])
  plt.xticks([]), plt.yticks([])

plt.show()
```
